# **CSV TO XML**

```python
import xml.etree.ElementTree as ET
import pandas as pd
from json import loads, dumps
```

```python
values_df = pd.read_excel(save_path + "XML_test_flat.xlsx")
values_df.head()
```

```python
address_df = pd.read_excel(save_path + "XML_test_flat.xlsx", sheet_name=1)

address_df = address_df[values_df.columns] #reorder address_df based on values_df
address_df.head()
```

```python
values_dict = loads(values_df.to_json(orient="split"))
print(values_dict)
```

```python
address_dict = loads(address_df.to_json(orient="split"))
print(address_dict)
```

```python
def create_root(address_list, value, root):

    if(len(address_list) == 1): #if end of address_list, add value to new_root with address id
      new_root = ET.SubElement(root, address_list[0])
      new_root.text = value
      return

    elif root.find(address_list[0]) != None: #else if existing new_root in address_list[0], go inside new_root
      create_root(address_list[1:], value, root.find(address_list[0]))

    else: #else if no such address exists in address_list[0], create new_root
      new_root = ET.SubElement(root, address_list[0])
      create_root(address_list[1:], value, new_root)
```

**Output 1 Type - Each Entry under each new Rpt XPath**

```python
root = ET.Element('XML_test_flat') #name root 'Trade_data'

for row in range(len(address_dict['index'])): #each row entry in excel

  values_list = values_dict['data'][row]

  row_entry_root = ET.SubElement(root, "Rpt") #new_root of each row entry, corresponding to first id of address_list

  for col in range(len(values_list)): #each column entry in each row, -1 as first id used to plot new_root of each row entry

    address_list = address_dict['data'][row][col].split('/')[1:] #get address of each value
    address_list.insert(0, "New")

    value = str(values_list[col])

    create_root(address_list, value, row_entry_root) #create new roots to insert value using the new_root for each row entry
```

```python
ET.indent(root, space="   ", level = 0)
ET.dump(root)
```

```python
tree = ET.ElementTree(root)
tree.write(save_path + "Output1.xml", encoding='utf-8', xml_declaration=True)
```

**Output 2 Type - All entries sorted by XPath**

```python
root = ET.Element('XML_test_flat') #name root 'Trade_data'

for row in range(len(address_dict['index'])): #each row entry in excel

  values_list = values_dict['data'][row]

  for col in range(len(values_list)): #each column entry in each row, -1 as first id used to plot new_root of each row entry

    address_list = address_dict['data'][row][col].split('/')[1:] #get address of each value
    address_list.insert(0, "New")
    address_list.insert(0, "Rpt")

    value = str(values_list[col])

    create_root(address_list, value, root) #create new roots to insert value using the new_root for each row entry
```

```python
ET.indent(root, space="   ", level = 0)
ET.dump(root)
```

```python
tree = ET.ElementTree(root)
tree.write(save_path + "Output2.xml", encoding='utf-8', xml_declaration=True)
```

```python
import json
import xmltodict
#import xml.etree.ElementTree as ET
import pandas as pd
from json import loads, dumps
from google.colab import files

# open the input xml file and read data in form of python dictionary using xmltodict module
xml_file = open(save_path + 'FX_MAS_NEWT_OTHR.xml')
data_dict = xmltodict.parse(xml_file.read())
xml_file.close()

print(data_dict)
```

```python
tag = 'TradData'

def get_tag_data(data_dict, tag):

  if tag in data_dict:
    #print("tag found:", data_dict)
    return data_dict

  else:
    result = None
    #go into all keys of current data_dict
    for key in data_dict:
      if(isinstance(data_dict[key], str)): #check if current_data_dict has only 1 string (ie not a dict key), continue if so
        continue

      result = get_tag_data(data_dict[key], tag) #go into the dict[key]

    return result

data = get_tag_data(data_dict, tag)
print(data)

trad_data_dict = data[tag]
print(trad_data_dict)
```

```python
df = pd.json_normalize(trad_data_dict, record_path = ['Rpt']) #record_path takes list
df.head()
```

```python
import xml.etree.ElementTree as ET
import pandas as pd
from google.colab import files

values_df = pd.DataFrame()
address_df = pd.DataFrame()

my_file = open(save_path + 'FX_MAS_NEWT_OTHR.xml')
parser = ET.XMLPullParser(['start', 'end'])
parser.feed(my_file.read())

stack = [] #create stack for xml hierarchy
row = [] #create row of values
row_address = [] #create row_address of xml addresses

for event, elem in parser.read_events():

    if(elem.tag[elem.tag.index('}')+1:] not in stack):
      stack.append(elem.tag[elem.tag.index('}')+1:])
    else:
      stack.remove(elem.tag[elem.tag.index('}')+1:])
    #print(stack)

    if('TradData' not in stack): #only get data from TradData hierarchy
      continue

    if((len(elem.text) > 1) and (event == 'start')): #check if text is empty and start event

      #print("stack:", stack)
      row.append(elem.text)
      row_address.append('/'.join(stack))

    if((event == 'end') and (elem.tag[elem.tag.index('}')+1:] ==  'Rpt')): #check for new row using Rpt and end event

      if(len(row) != 42): #most rows do not have PrrTxId, hence add empty string
        row.insert(17, '')
        row_address.insert(17, '')

      values_df = pd.concat([values_df, pd.DataFrame([row])],axis=0, ignore_index=True) #concat row and row address
      address_df = pd.concat([address_df, pd.DataFrame([row_address])],axis=0, ignore_index=True)

      row = [] #clear row and row address
      row_address = []


values_df.columns = ['RptgCtrPty_LEI','RptgCtrPty_TradgCpcty','RptgCtrPty_CtrPtySd','RptgCtrPty_TradrLctn', 'RptgCtrPty_BookgLctn', 'OthrCtrPty_LEI','SubmitgAgt_LEI','ClrMmb_LEI', 'NttyRspnsblForRpt_LEI', 'RptgTmStmp', 'CtrctTp', 'AsstClss', 'UnqPdctIdr', 'ISIN',
                                  'SttlmCcy_Ccy', 'SttlmCcyScndLeg_Ccy', 'TxId_UnqTxIdr', 'PrrTxId_UnqTxIdr', 'Cd', 'PltfmIdr', 'FrstLeg_@Ccy', 'ScndLeg_@Ccy',  'DlvryTp', 'ExctnTmStmp', 'FctvDt',
                                  'XprtnDt', 'SttlmDt', 'DerivEvt_Tp', 'DerivEvt_DtTm', 'TradConf_Confd_Tp', 'TradConf_Confd_TmStmp', 'TradClr_LEI', 'TradClr_ClrDtTm', 'XchgRate', 'BaseCcy', 'QtdCcy', 'OthrPmt_Amt_@Ccy', 'PmtTp_Tp',
                                  'PmtPyer_LEI','PmtRcvr_LEI', 'CmplxTradId', 'TechRcrdId']

address_df.columns = ['RptgCtrPty_LEI','RptgCtrPty_TradgCpcty','RptgCtrPty_CtrPtySd','RptgCtrPty_TradrLctn', 'RptgCtrPty_BookgLctn', 'OthrCtrPty_LEI','SubmitgAgt_LEI','ClrMmb_LEI', 'NttyRspnsblForRpt_LEI', 'RptgTmStmp', 'CtrctTp', 'AsstClss', 'UnqPdctIdr', 'ISIN',
                                  'SttlmCcy_Ccy', 'SttlmCcyScndLeg_Ccy', 'TxId_UnqTxIdr', 'PrrTxId_UnqTxIdr', 'Cd', 'PltfmIdr', 'FrstLeg_@Ccy', 'ScndLeg_@Ccy',  'DlvryTp', 'ExctnTmStmp', 'FctvDt',
                                  'XprtnDt', 'SttlmDt', 'DerivEvt_Tp', 'DerivEvt_DtTm', 'TradConf_Confd_Tp', 'TradConf_Confd_TmStmp', 'TradClr_LEI', 'TradClr_ClrDtTm', 'XchgRate', 'BaseCcy', 'QtdCcy', 'OthrPmt_Amt_@Ccy', 'PmtTp_Tp',
                                  'PmtPyer_LEI','PmtRcvr_LEI', 'CmplxTradId', 'TechRcrdId']
```
